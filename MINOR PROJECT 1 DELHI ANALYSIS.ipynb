```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('delhi datewise data.csv')
```

```python
df
```

```python
df.head()
```

```python
plt.figure(figsize=(14,8))
plt.bar(df.Date,df.New_cases,label='New_cases',color='purple')
plt.title("covid_analysis_june_july:delhi")
plt.xticks(df.Date,rotation=90)
plt.xlabel("CASES DAILY (15june-15july)",labelpad=12)
plt.ylabel("Number of Cases ",labelpad=12)
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.bar(df.Date,df.New_cases,label='New_cases',color='blue')
plt.bar(df.Date,df.New_recovered,label='New_recovered',color='green')
plt.bar(df.Date,df.New_deaths,label='New_deaths',color='red')
plt.legend()
plt.title("covid_analysis_june_july_delhi")
plt.xticks(df.Date,rotation=90)
plt.xlabel("FROM 15JUNE-15JULY",labelpad=18)
plt.ylabel("no of cases",labelpad=18)
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')

plt.figure(figsize=(14,8))
plt.plot(df.Date,df.New_cases,label='New_cases',color='blue',linewidth=3)
plt.plot(df.Date,df.New_recovered,label='New_recovered',color='green',linewidth=5)
plt.plot(df.Date,df.New_deaths,label='New_deaths',color='red',linewidth=3)

plt.legend()
plt.title("covid_analysis_june_july_delhi")

plt.xticks(rotation=90)
plt.xlabel("Daily cases",labelpad=20)
plt.ylabel("Number of cases",labelpad=5)
plt.grid(True,color='k')

plt.show()
```

```python
df.hist()
```

```python
plt.pie(df.New_recovered, labels=df.Date)
```

```python
df.plot(y=['New_recovered','New_cases','New_deaths'],x="Date",rot=90)
```

```python
labels=['confirmed','recovered','deceased']
sizes=[131219,116372,3853]
colors = ['pink', 'powderblue','red']
explode = (0, 0, 0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, center=(0.8,0.8), startangle=0)
plt.axis('equal')
plt.show()
```

```python
IFR_DATEWISE=(df['New_recovered']/df['New_deaths'])*100
```

```python
plt.figure(figsize=(14,8))
plt.bar(df.Date,IFR_DATEWISE,label='New_recovered',color='orange')
plt.bar(df.Date,df.New_deaths,label='New_deaths',color="darkblue")
plt.title("COVID-19 DELHI_IFR")
plt.xticks(df.Date,rotation=90)
plt.xlabel("Date",labelpad=10)
plt.ylabel("IFR",labelpad=10)
plt.show()
```

```python
df.plot.bar(y=['New_cases','New_recovered','New_deaths'], x='Date')
```

```python
df.plot(y=['New_cases','New_recovered','New_deaths'],x="Date",rot=90)
```
